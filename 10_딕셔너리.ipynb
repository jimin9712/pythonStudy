# **1. 딕셔너리(Dictionary)**

- 대응관계를 나타내는 자료형으로  key와 value라는 것을 한 쌍으로 갖는 형태
- 하나의 딕셔너리의 key는 중복될 수 없음
- 하나의 딕셔너리의 value는 중복될 수 있음

### 1-1. 딕셔너리 만들기

```python
dic = {}
print(type(dic))
```

```python
dict1 = {1:'이지민',2:'이유진',3:'용호중',4:'이세찬'}
print(dict1)
```

### 1-2. key를 통해 value 찾기

```python
dict1 = {1:'이지민',2:'이유진',3:'용호중',4:'이세찬'}
# print(dict1[1])  대괄호 사이에 인덱스 번호는 절대 적지 않는다.
print(dict1[3]) # key 값을 통래 value 가져오기!
```

```python
#key값을 같은 자료형으로 해줄 필요는 없다.
user1 = {'no':1,'user_id':'Jimin','name':'이지민',4:22}
print(user1['no'])
print(user1['user_id'])
print(user1['name'])
print(user1[4])
```

### 1-3. 데이터 추가 및 삭제

```python
foods = {1:'rice'}
print(foods)
#추가
#데이터를 추가하는 문법을 사용해서 추가도 할 수 있다.
# 해당 딕셔너리에 없는 키 값을 저장공간으로서 사용한다면
# 키 값을 하나 만들어서 데이터를 저장한다.
foods[2] = 'chicken'
print(foods)

foods[2] = '잡채'
print(foods)

#삭제
del foods[2]
print(foods)
```

### 1-4. 딕셔너리 함수

```python
user1 = {'no':1,'user_id':'Jimin','name':'이지민',4:22}

# keys() : key 리스트를 반환
print(user1.keys())
print(type(user1.keys()))

# values() : value 리스트를 반환
print(user1.values())

#items() : key 와 value를 한 쌍으로 묶는 튜플을 반환
print(user1.items())

print('\n-----get-----')
#get() : key를 이용해서 value를 반환
#print(user1['id']) # 해당 딕셔너리에 키 값이 없는 딕셔너리를 찾는다면 에러!
print(user1.get('name'))
print(user1.get('이름')) # 없는 키를 전달하면 에러가 아니느 None을 반환!
print(user1.get('이름', '해당 키 값이 존재하지 않습니다')) # 옵션 추가 시 해당 데이터를 반환.


print('\n-----in-----')
# in : key가 딕셔너리 안에 있는가를 검사. 논리타입의 값
print('name' in user1)
print('이름' in user1)
```

### 1-5. 딕셔너리와 for문

```python
user1 = {'no':1,'user_id':'Jimin','name':'이지민',4:22}

for i in user1:
  print(i, end=' ') # 키만 가지고 나온다
print('\n')
for i in user1.keys():
  print(i, end = ' ')# 키만 가지고 나온다
print('\n')
for i in user1.values():
  print(i, end = ' ')# 값만 가지고 나온다
print('\n')
for i in user1:
  print(user1[i], end = ' ')
print('\n')
for i in user1:
 print(user1.get(i), end = ' ')
```

### 문제
아래와 같은 딕셔너리 자료에서 주민등록번호를 추출하여 성별을 알아내는 프로그램을 작성해보자.

> user = {'name' : '이지민', 'hp' : '010-1111-0000', 'ssn' : '0208291914578'}

```python
user = {'name' : '이지민', 'hp' : '010-1111-0000', 'ssn' : '0208291914578'}

if 'ssn' in user:
  ssn = user.get('ssn')
  if ssn[6] == '1' or ssn[6]=='3':
    print('남')
  elif ssn[6] == '2' or ssn[6]=='4':
    print('여')
  else:
    print('제대로 불러오지 못했어요')
else:
  print('ssn이 없습니다.')
```
