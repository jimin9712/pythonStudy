### 1-3. 생성자
- 클래스 인스턴스가 생성될 때 자동으로 호출
-  __ init __(self, 변수1, 변수2, ...)
- self 매개변수는 항상 첫번째 오며, 자기 지신의 객체를 가리킴
-  이름이 꼭 self 일 필요는 없지만 관례적으로 self를 사용
- 생성자에서는 해당 클래스가 다루는 데이터를 정의

````python
# self가 진짜 위치 정보가 넘어 오는 지를 확인
class Dog :
  def __init__(self): # self는 객체 그 자체!
    print(self, 'init 호출')
````

````python
jin = Dog() # 이거는 미리보기 아님
print(jin)

tomy = Dog()
print(tomy)
````

````python
# 클래스가 다루는 데이터를 정의
class Dog :
  def __init__(self):
    print(self, 'init 호출')
    self.name = '이름 없음'
    self.age = 0
````

````python
#객체 만들기 / 생성자가 그대로 실행이 됨 self 위치 정보 안에 name 이라는 걸 만들어서 그 안에 넣어놔라.
jin = Dog()
print(jin.name) # 값

jin.name = '진순이'
print(jin.name)

tomy = Dog()
print(tomy.name)
````

````python
# 객체를 만듦과 동시에 내가 원하는 데이터를 필드에 넣기
class Dog :
  def __init__(self,name,age, breed = '진돗개'):
    print(self, 'init 호출')
    self.name = name
    self.age = age
    self.breed = breed
````

````python
jin = Dog('진순이', 1)
print(jin.name)
print(jin.breed)

tomy = Dog('이유진', 25, '최고')
print(tomy.name)
print(tomy.age)
print(tomy.breed)

# 함수 호출과 똑같이 매개변수를 잘 넘겨줘야합니다.
# self 자동으로 넘어가기에 객체를 만들 때 신경 쓸 필요가 없다.
````

### 1-4. 메소드 정의하기
- 해당 클래스의 **객체에서만 호출 가능**한 함수
- 해당 객체의 속성에 대한 연산을 행함
- **객체이름.메소드명()** 형태로 호출함

````python
class Counter:
  # 생성자
  def __init__(self):
    self.num = 0

  # 메소드 해당 객체의 기능
  # 번호표 뽑았다! 증가!
  def increment(self):
    self.num += 1

  # 번호표 리셋!
  def do_reset(self):
    self.num = 0

  # 공통적인 기능을 가지고 있는 친구들을 묶어줄 때
  # add (수학 연산과 관련된 함수들을 묶어줄 때)
  def add(self, num1, num2):
    return num1 + num2
````

````python
SHBank = Counter()
HNBank = Counter()

print('신한 번호표 ', SHBank.num)
print('신한 번호표 ', HNBank.num)

# 번호표를 증가한 객체는 신한은행 객체!
# 그렇기 때문에 하나은행 객체는 아무 영향을 받지 않는다.
SHBank.increment()

print('신한 번호표 ', SHBank.num)
print('신한 번호표 ', HNBank.num)

SHBank.do_reset()
print('신한 번호표 ', SHBank.num)
print('신한 번호표 ', HNBank.num)
````

### 1-5. 메소드 타입
- instance method : 객체 형태로 호출되기 때문에 해당 메소드를 호출한 객체에만 영향을 미침
- class method : class로 호출(함수 선언 위에 @staticmethod 라고 표시)

````python
class Math:
  def add(self, x, y):
    return x + y
  def multiplication(self, x,y):
    return x * y
````

````python
math = Math()
print(math.add(5,3))
print(math.multiplication(4,3))
````

````python
class Math2:
  # 어노테이션 문법
  # 아래 영역에 쓰임을 컴퓨터에게 알려주는 용도
  @staticmethod
  def add(x, y):
    return x + y

  @staticmethod
  def multiplication(x,y):
    return x * y
````

````python
# 메소드의 소속만 알려주면 된다.
# 객체를 만들 필요 없이 사용 가능
print(Math2.add(5,3))
print(Math2.multiplication(5,4))
````
