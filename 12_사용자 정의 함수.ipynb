함수명(값1,값2...) # 호출

````python
#fun2(49) 이 형태로 호출이 되었다면
#num = 49 가 실행이 된 후
#함수 내에 선언 되어있는 코드가 실행이 된다.
def func2(num):
  print(f'매개변수로 넘어온 값은 {num} 이다.')
````

````python
func2(49)

#변수의 범위, 스코프
# print(num) #매개변수는 지역변수라서 해당 영역 내에서만 사용 가능.
````

````python
# 매개변수를 설정해놨는데 넘겨주지 않는다면 에러가 발생한다.
# 발생하지 않는 경우도 있다. 그건 밑에 디폴트 매개변수 참고!
# func2()
````

````python
# 두 정수를 매개변수로 받아서
# start 부터 end 까지 정수의 합을 구하는 함수

def addAll(start,end):
  sum = 0
  for i in range(start, end + 1):
    sum += i
  print(f'{start}부터 {end}까지의 총 합은 {sum}입니다.')
````

````python
addAll(1,10)
````

````python
# 매개변수는 기본적으로 작성해준 순서대로 들어가게 된다.
# 순서대로 넣고 싶지 않다면 매개변수명을 명시해준다.
addAll(end = 100, start = 1)
````

### 1-3. 반환(return)값이 있는 함수

```
def 함수명(변수1, 변수2,...):
    함수가 호출되면 실행할 문장
    ...
    return 값
```
```
함수명(값1,값2...) # 호출
```

````python
def printMind():
  return '빨리 설날이 왔으면 좋겠다.'
````

````python
#리턴이 있는 함수는
#함수가 호출된 부분 전체를 리턴 값으로 볼 줄 알아야 한다
temp =  printMind()
print(temp)
````

````python
# start에서 end까지 정수의 합을 리턴!
def addAll_re(start,end):
  sum = 0
  for i in range(start, end +1):
    sum += i
  return sum
````

````python
start, end = map(int, input('시작과 끝을 스페이스바로 구분').split(' '))

result = addAll_re(start, end)
print(f'{start}부터 {end}까지의 총 합은 {result}입니다.')
````

### 1-5. 매개변수의 개수가 가변적인 함수

```
def 함수명(*변수):
    함수가 호출되면 실행할 문장
    ...
    return 값
```
```
변수 = 함수명(값1, 값2, ...) # 호출
변수 = 함수명() # 호출
```

````python
def add(*args):
  print(args)
  print(type(args))
  sum = 0
  for i in args:
    sum += i
  return sum
````

````python
print(add(10,20,30))
````

````python
# 리스트 타입의 값을 넘겨서 실행!
def add(nums):
  print(nums)
  print(type(nums))
  sum = 0
  for i in nums:
    sum += i
  return sum
````

````python
ex = [10,20,30,40]
print(add(ex))
````

### 1-7. 리턴값이 여러 개인 함수

```
def 함수명(변수1, 변수2,...):
    함수가 호출되면 실행할 문장
    ...
    return 값1, 값2, ...
```
```
변수1, 변수2, ... = 함수명(값1, 값2, ...)# 호출
```

````python
def getResult(num1, num2):
  return num1+num2 , num1 - num2, num1 * num2, num1 / num2
````

````python
print(getResult(5,7))
print(type(getResult(5,7)))
````

````python
#튜플의 원소를 각각 저장.
plus, minus, multiple, divide = getResult(5,7)
print(plus)
print(minus)
print(multiple)
print(divide)
````

````python
#내가 사용하고 싶은 친구만 저장!
_, _, temp, _ = getResult(4,9)
print(temp)
````

### 문제
- 700원 짜리 음료수를 뽑을 수 있는 자판기 프로그램을 구현해보자.
- 매개변수로 금액을 입력!
- 아래 텍스트가 나오도록 구현.
- 출력 예시

```
vending_machine(3000) # 함수 호출

실행 결과
음료수 = 0개, 잔돈 = 3000원
음료수 = 1개, 잔돈 = 2300원
음료수 = 2개, 잔돈 = 1600원
음료수 = 3개, 잔돈 = 900원
음료수 = 4개, 잔돈 = 200원
```

````python
def vending_machine(money):
  price = 700
  drinks = 0

  while money >= price:
      drinks += 1
      money -= price
      print(f"음료수 = {drinks:2d}개, 잔돈 = {money:4d}원")
````

````python
vending_machine(3000)
````

````python
def vending_machine(money):
  count = money // 700
  for i in range(count + 1):
    change = money - 700 * i
    print(f'음료수 = {i:2d}개, 잔돈 = {change}')
````

````python
vending_machine(3000)
````
