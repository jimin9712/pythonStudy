# **1. 문자열 인덱싱**

- 문자열에 번호를 매겨 특정 문자를 찾을 수 있는 기능
- 인덱스 번호는 0부터 순차적으로 자동 부여된다.

```python
string ="문자열에 번호를 매겨 특정 문자를 찾을 수 있는 기능"
print(string)
```

```python
print(string[0])
print(string[5])
print(string[4]) # 공백 문자
print(string[-4])

# 변수의 개념 조금 더 숙지!
# 값과 저장공간의 차이!
print('이지민'[2])
```

# **2. 문자열 슬라이싱**

- 문자열에서 원하는 부분을 추출하는 기능

```python
string2 = "문자열에서 원하는 부분을 추출하는 기능"
print(string2)
```

```python
print(string2[0:4]) # index 0부터 4 이전까지의 문자열을 잘라서 가져옴
print(string2[0:4][0])
print(type(string2[0:4][0]))
print(string[:4]) #처음부터 4가 되기 직전까지!
print(string2[10:]) #10부터 끝까지!
```

# **3. 문자열 함수**

- 문자열과 관련된 여러가지 기능(함수)

```python
string3 = "문자열 함수는 문자열과 관련된 여러가지 기능"
print(string3)
```

```python
#count() : 문자열에서 매개변수로 전달된 문자열의 출현 횟수를 반환
print(type(string3.count('문')))
print(string3.count('문'))
print(string3.count('문') + 5)
print(string3.count('이지민')) # 없으니까 0
```

```python
#find() : 매개변수로 전달된 문자열이 처음 나오는 위치를 인덱스로 반환
print(string3.find('문'))
print(string3.find('류') ) # 없으면 -1을 반환한다.
print(string3.find('함수')) # 문자열도 되긴 되는데 쓰는 목적이 다르다.
```

```python
# 플러스 알파
print(string3.find('문'))
first = string3.find('문',5)
print(string3.find('문', first + 1))
```

```python
# index() : 매개변수로 전달된 문자가 처음 나오는 위치를 인덱스로 반환
print(string3.index('열'))
print(string3.index('라')) #ValueError
```

```python
#replace() : 특정 문자열을 치환
#변수 안의 값을 수정하는 것이 아닌, 바꾼 값을 반환해준다.
print(string3)
print(string3.replace('기능','함수'))
#값의 수정
string3 = string3.replace('기능','함수')
print(string3)
```

```python
# 공백 제거
# lstrip() : 문자열의 왼쪽 공백을 제거
# rstrip() : 문자열의 오른쪽 공백을 제거
# strip() : 문자열의 양쪽 공백을 제거
# 제거한 후 반환!
string4 = '         외계인2부                  썬더            '
print(string4)
print(string4.lstrip())
print(string4.rstrip())
print(string4.strip())
#문제 : 모든 공백을 제거하기
print(string4.replace(' ', ''))
```
