```python
print('미리 인사드립니다.\n주말 잘 보내세요')
#tab, 일정 간격 띄어쓰기
print('미리 인사드립니다.\t주말 잘 보내세요')
#\로 문자 자체를 죽이는 작업
print('나는 너에게 \'사랑해\'라고 말했다.')
print("나는 너에게 '사랑해'라고 말했다.") #이건 된다.
```

# **3. 출력 서식 지정하기**

1. %d(정수), %f(실수), %s(문자열)를 이용하여 출력 서식을 지정
2. 서식을 제외한 문자는 입력한 그대로 출력이 됨
3. 서식을 지정해 출력할 데이터가 2개 이상인 경우(소괄호)로 데이터를 묶어야 함

```python
print('숫자 : 100')
print('숫자 : %d' %100)
print('숫자 : %d' %100.5)
```

```python
print('실수 : %f' % 100.5)
```

```python
print('문자열 : %s'%'이지민!')
```

```python
print('10 + 5 = 15')
print('%d + %d = %d' %(10,5,10+5))
```

```python
# %d, 자릿수 설정
print('%02d' % 9)
print('%d' % 10)
#데이터의 값이 달라지기 때문에 안된다.
print('%5d' % 10000000)
```

```python
# %f, 자릿수 설정
print('%05.2f'%3.141592)
print('%.4f'%3.141592) # 반올림
```

```python
# %s, 자릿수 설정
print('%8s'%'쉬는시간')
```

# **4. format 함수와 출력 인덱스 사용**

```python
#인덱스(index) : 0부터 순서대로 진행되는 점수
print('{} {}'.format('이번 주는 금요일에', '강의 있어요!'))
print('이번 주는 {0}가 {1}있어요!'.format('강의', '잡혀'))
print('이번 주는 {1}가 {0}있어요!'.format('강의', '잡혀'))
```

```python
print('{:5s}님은 {:03d}살 입니다.'.format('이지민',25))

#f-stronfs
#서식을 지정할 수 있고, 성격이 비슷한 친구
#변수를 배운 후에 알아보도록 하자!\
#변수.ipynb 참고!
```
