### 2-1. if문

````python
# 조건을 만드는 방법
# 비교(관계)
# 연산된 결과가 항상 True 나 False로 나온다
# <,>,<=,>=,==,!=
print(10 > 5)
print(10 == 10)
print(10 != 10)
````

- if문 기본 문법

```
if문 조건식(결과가 True 또는 False가 나오는 문장):
  조건식이 True인 경우 실행할 문장
```

````python
# 나이를 입력 받아 성인인지 아닌지 판단하는 프로그램

age = int(input('나이를 입력하세요 >>'))
#들여쓰기 신경쓰기
if age > 19:
  print('성인입니다.')
#조건식과 상관없이 무조건 실행되는 코드
# 들여쓰기가 되어있지 않기 때문에
# if문 영역 안에 있는 문장이 아니다.
print('입력한 나이 : ',age)
````

````python
age = int(input('나이를 입력하세요 >>'))

if age > 19: print('성인입니다.')

print('입력한 나이:' ,age)
````

### 2-2. if~else문
```
if 조건식(결과가 True 또는 False가 나오는 문장 ):
  조건식이 True인 경우 실행할 문장
else :
  조건식이 False인 경우 실행할 문장
```

````python
age = int(input('나이를 입력하세요 >> '))

if age > 19 :
  print('성인입니다.')
else :
   print('나가라')
print('입력한 나이:',age)
````

### 2-3. if ~ elif ~ else 문
```
if 조건식1 (결과가 True 또는 False가 나오는 문장 ):
  조건식이 True인 경우 실행할 문장
elif 조건식2 :
  조건식2가 True인 경우 실행할 문장
elif 조건식3 :
  조건식3가 True인 경우 실행할 문장
...
else :
  조건식 모두가 False인 경우 실행할 문장
```
- elif에 적은 조건식은 위에 조건식이 False일 때만 검사한다.
- 즉, 위에 조건식이 True라면 아래 조건들은 검사조차 하지 않는다.
- 위에 조건식의 결과에 상과없이 전부 다 검사하고 싶다면 elif 대신 if를 써야한다.



````python
age = int(input('나이를 입력하세요 >> '))

if age > 19 :
  print('성인입니다.')
elif age > 14:
   print('청소년입니다.')
elif age > 6:
   print('어린이입니다.')
elif age > 0:
   print('아기입니다')
else :
  print('잘못 입력하셨습니다')
print(f'입력한 나이:{age}')
````

````python
# elif 를 사용하지 않고 if만 사용해보기
# 위에 조건식의 여부와 상관없이 전부 검사!
# 전부 독립적인 if문이기 때문!
# 문법상 전혀 문제가 없는 코드

# 단, 지금은 서비스상 흐름이 잘못된 코드라고 볼 수 있다.
age = int(input('나이를 입력하세요 >> '))

if age > 19 :
  print('성인입니다.')
if age > 14:
   print('청소년입니다.')
if age > 6:
   print('어린이입니다.')
if age > 0:
   print('아기입니다.')
else :
  print('잘못 입력하셨습니다')
print(f'입력한 나이:{age}')
````

### 2-4. if문 한 줄로 표현하기 (like 삼항연산자 참고)
- (True일 때 코드 ) if 조건식 else (False일 때 코드)
- 조건식이 하나만 있을 때 한 줄만에 끝낼 수 있다.
- 만약, 실행 코드 부분에 값이 들어가 있다면 if문 전체를 하나의 값으로도 볼 수 있다.

````python
data = int(input('숫자를 입력하세요 >> '))

# print('입력한 숫자는 5보다 크다.') if data > 5 else print('입력한 숫자는 5보다 작다.')
# print('입력한 숫자는 5보다 크다.') if data > 5 else print('입력한 숫자는 5랑 같다.') if data == 5 else print('입력한 숫자는 5보다 작다.')
result = '크다' if data > 5 else 1 if data == 5 else '작다'
print(result)

result2 = print('입력한 숫자는 5보다 크다.') if data > 5 else print('입력한 숫자는 5보다 작다.')
print(result2)
````

### 문제
- 국어,영어,수학 점수를 입력받아 총점, 평균, 학점을 구하는 프로그램을 작성해보자.
```
단, 학점은 평균 기준이며
90점 이상은 A학점,
80점 이상은 B학점,
70점 이상은 C학점,
60점 이상은 D학점,
나머지는 F학점으로 한다.
```

````python
kor,eng,math = map(int,input('국어, 영어, 수학점수를 /를 기준으로 나눠서 입력하세요 >>').split('/'))
total = kor + eng + math
mean = total/3
if mean >=90 :
  grade ='A'
elif mean >=80 :
   grade ='B'
elif mean >=70 :
  grade ='C'
elif mean >=60 :
  grade ='D'
else :
  print('F학점')
print(f'당신의 총점은{total}이고,\n당신의 평균 점수는{mean}입니다.\n학점은{grade}입니다.')
````

### 문제 2.
- 숫자를 입력받아 2의 배수인지, 3의 배수인지, 아니면 2,3의 배수가 모두 아닌지 판별하는 프로그램을 작성해보자.

````python
num = int(input('숫자를 입력하세요 >>'))

if num % 6 == 0 :
  print(f'{num}은 2와 3의 공배수')
elif num % 2 == 0 :
  print(f'{num}은 2의 배수')
elif num % 3 == 0 :
  print(f'{num}은 3의 배수')
else :
  print('둘 다 아녀')
````

### 2-5. 논리 연산자(and)
 - 양쪽 두 조건의 값이 모두 True인 경우 결과가 True가 되는 연산자

| A | B | 결과 |
|-----|-----|-----|
|True|True|True|
|True|False|False|
|False|True|False|
|False|False|False|

````python
#관리자 로그인 프로그램 만들ㄹ기
#아이디: admin, 비밀번호 : 1234

user_id = input('아이디 >>')
user_pw = input('비밀번호 >>')

if user_id == 'admin' and user_pw == '1234':
  print('관리자가 로그인 하였습니다.')
else:
  print('아이디와 비밀번호를 다시 확인하세요!')
````

### 2-6. 논리연산자(or)
- 양쪽 두 조건에 값이 하나라도 True인 경우 결과가 True가 되는 연산자

| A | B | 결과 |
|-----|-----|-----|
|True|True|True|
|True|False|True|
|False|True|True|
|False|False|False|

````python
# 주민등록번호로 성별 판별
serialNo = input('주민등록번호 13자리를 -없이 입력하세요!')

# 인덱스 번호 잘 설정했는지 확인!
# print(serialNo[6])

if serialNo[6] == '1' or serialNo[6] == '3':
  print('남자에요!')
elif serialNo[6] == '2' or serialNo[6] == '4':
  print('여자에요!')
else :
  print('다시 확인하세요!')
````

# **참고**
- 연산자 우선순위
  - 최우선 -> ( )
  - 산술 -> +, -, *, / ...
  - 관계 -> <, >, <=, >=, ==, !=, ...
  - 논리 -> and, or
  - 대입 -> =

- 결합성
  - 같은 우선 순위의 연산자들은 왼쪽부터 연산이 이루어진다.
