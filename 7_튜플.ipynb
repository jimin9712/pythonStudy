# **1. 튜플(Tuple)**
- 리스트와 비슷하지만 리스트는 [ ]로 자료를 감싸고, 튜플은 ( )로 감싼다.
- 수정과 삭제가 불가능

```python
tu1 = ()
print(type(tu1))
print(tu1)

tu2 = (1,2,3,4)
print(tu2)
#대괄호든 , 소괄호든, 생략하면 튜플
tu2_1=1,2,3,4
print(tu2_1)
#튜플도 여러가지 자료형을 담을 수 있다.
tu3 = (1,3.14,'이지민',True,('된다.','안된다.'))
print(tu3)
tu5 = (1,3.14,'이지민',True,['된다.','안된다.'])
print(tu5)

#인덱싱과 슬라이싱 리스트와 전부 동일
print(tu5[-1])
print(tu5[0:3])
```

# **2. 튜플의 수정 및 삭제**

```python
tu6 = (1,2,'apple','orange')
print(tu6)
#수정
#튜플은 수정이 불가능하다
#tu[1] = 49

#삭제
#튜플은 삭제도 불가능!
#del tu6[1]
#튜플 자체는 삭제 가능 / 요소의 삭제가 안되는 것
del tu6
```

```python
tu7 = (1,3.14,'이지민',True,['된다.','안된다.'])

#이건 되는 이유 : 1,3.14,'이지민',True, 다음엔 주소가 저장되기 때문에 <6_리스트>에 얕은 복사 참조
# 튜플의 요소를 수정하고 있는게 아니다.
# 튜플의 -1번째 인덱스는 값이 아니라 주소가 요소이다.
# 주소를 바꾸는 게 아니라 그 주소 안에 있는 리스트의 요소를 바꾸는 것
#그렇기 때문에 가능!
tu7[-1][-1] = '수정'
print(tu7)
print('tu7의 내부 리스트 주소 :',id(tu7[-1]))

#append를 한다고 해서 주소가 달라지지 않기 때문에!
tu7[-1].append('가능')
print('tu7의 내부 리스트 주소 :',id(tu7[-1]))
print(tu7)
#tu7.append(10) 튜플은 추가 기능 자체가 없다.
```

# **3. 튜플의 연산**

```python
tu8 = (1,2,'이지민','이유진')
tu9 = (49,66)
list1 = [49,66]
#덧셈 == 연결
print(tu8 + tu9)
#리스트와 튜플 연산은 불가

tu10 = tuple(list1)
print(tu10)

print('\n\n튜플과 리스트의 덧셈을 할 때 타입을 맞춰주자.')
print(tu9 + tuple(list1))

#곱셈(반복)
print(tu9 * 3)

#요소의 갯수
print(len(tu9))
```

# **4. 튜플의 정렬**

```python
tu11 = (1,3,2,0)
#tu11.sort()  sort()는 튜플 클래스 안에 존재하지 않음

print(sorted(tu11))
```
