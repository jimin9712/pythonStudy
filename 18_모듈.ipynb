# **1. 모듈**
- 클래스 또는 함수를 모아놓은 파일
- 모든 파이썬 파일(.py)은 모듈

```
# 모듈의 전체 함수 가져오기
import 모듈명

# 사용
모듈명.함수명()
```

````python
import test
````

````python
print(test.total(1,10))
````

```
#특정 함수만 골라서 가져오기
from 모듈명 import 함수 1, 함수 2, ...

# 사용
함수명()
```

````python
from test import calc
````

````python
_,_, result, _ = calc(4,9)
print(result)
````

```
#특정 함수만 골라서 가져오기
from 모듈명 import * (all)
# 사용
함수명()
```

````python
from test import *
````

````python
print(calc(1,10))
print(total(1,10))
````

```
# 모듈 별명 붙이기
import 모듈명 as 별명

# 사용
별명.함수명()
```

````python
import test as ts
ts.total(1,100)
````
