# **1. scope**

- 변수에 접근할 수 있는 범위
- local : 가장 가까운 함수 안 범위
- global : 함수 바깥의 변수 또는 import 된 module

```python
num1 = 10

def func1():
  num2 = 20
  print(num2)
  # 함수 호출이 종료되면 함수를 실행할 때 필요했던 변수들은 전부 제거.
```

```python
print(num1)
# print(num2 ) #num2의 스코프는 local! 함수까지의 범위
func1()
# print(num1) #함수 호출이 끝나면 해당 local 변수는 다 삭제!
```

```python
# locals() : 로컬 변수를 확인하는 함수
# globals() : 글로벌 변수를 확인하는 함수

num1 = 10

def func2():

  num2 = 20
  print(num2)
  print('num1이 로컬 변수가 맞아? :', 'num1' in locals())
  print('num1이 글로벌 변수가 맞아? :', 'num1' in globals())
```

```python
func2()
```

```python
num1 = 10

def func3():
  num1 = 100 # 로컬 변수 num1을 새롭게 만드는 작업.
  num2 = 20
  print(num2)
  print('num1이 로컬 변수가 맞아? :', 'num1' in locals())
```

```python
func3()
print(num1) # 함수 안의 num1은 로컬 변수! 글로벌 변수에는 영향을 주지 않는다.
```

# **2. global 키워드**
- 함수 내부에서 로컬 변수가 아닌 글로벌 변수로 사용하게 함

```python
wallet = 100000

def buy_coffee ():
  wallet -= 4500 # wallet = wallet - 4500
```

```python
print(wallet)
```

```python
wallet = 100000

def buy_coffee ():
  global wallet # 이 코드를 실행한 이 후부터 wallet은 global 변수로 사용된다.
  wallet -= 4500 # wallet = wallet - 4500
```

```python
print(wallet)
buy_coffee()
print(wallet)
```
