# **1. 변수**

- 변수란 처리할 데이터를 저장시키는 기억장소

```python
# = : 대입 연산자
# 대입연산자 앞에 변수가 있다면,
# 해당 변수는 저장공간으로써 쓰인다.
name = '이지민'
age = 25
#그 외에는 전부 다 저장되어 있는 값으로 쓰임
print(name)
print(age)

#이미 선언된 변수가 있다면 값을 덮어쓴다.
name = '이유진'
print(name)
```

```python
num1 = 10
num2 = 3
print('%d / %d = %.2f' %(num1, num2 , num1/num2))
```

- 영문자, 숫자, 특수문자(_)로 사용하며 대소문자를 구별

# **3. 변수의 데이터 타입**

- 데이터 타입(자료형) : 자료가 저장되는 형태를 의미

```python
name ='강감찬'
age = 25
weight = 65.5
isLover = True

#type() : 변수의 데이터 타입을 반환해주는 함수
print(type(name)) # <class 'str'> -> string, 문자
print(type(age)) # <class 'int'> -> 정수
print(type(weight)) # <class 'float'> -> 실수
print(type(isLover))# <class 'bool'> -> boolean 논리, 참, 거짓
```

# **4. f-strings**

- 파이썬 3.6 이후로 추가된 새로운 기능
- format() 메소드와 유사
- { } 를 사용하며 가독성이 뛰어나다는 장점이 있다.



```python
name = '이지민'
age = 25

print(f'{name}의 나이는 {age:03}살 입니다.')
```
