- 사용자에게 데이터를 입력받을 수 있는 함수
- 데이터를 문자형으로 입력


```python
age = input()
print(age)
print(type(age)) #입력한 데이터가 문자로 저장.
```

```python
name = input('이름을 입력하세요 >>')
print(f'{name}님! 멋진 이름이네요!')
```

```python
#split() : 매개변수로 전달된 문자를 기준으로 데이터를 나눠서 저장
data = '이지민/25/178'
name , age, height = data.split('/')
print(name,age,height)
```

```python
name, age , height = input('이름, 나이, 키를 /를 기준으로 나눠서 입력하세요 >>').split('/')
print(f'{name}님의 나이는 {age}살이며, 키는 {height}cm 입니다.')
```

```python
name, age , height = input('이름, 나이, 키를 공백을 기준으로 나눠서 입력하세요 >>').split(' ')
print(f'{name}님의 나이는 {age}살이며, 키는 {height}cm 입니다.')
```

```python
# type간의 산술 연산
#문자와 정수의 연산은 불가
#'123' + 5
'123' + '456' #문자와 문자의 연산은 연결
```

```python
#int() : 문자열을 숫자(정수)타입으로 변환
num1 = '100'
num2 = 50
print(int(num1) + num2)

num3 = '200'
num4 = '10'
print(int(num3 + num4))
```

### 문제
- 국어, 영어, 수학 점수를 입력 받아 총점, 평균을 구하는 프로그램을 작성해보자.

```python
kor,eng,math = input('국어, 영어, 수학점수를 /를 기준으로 나눠서 입력하세요 >>').split('/')
total = (int(kor)+ int(eng) + int(math))
mean = total/3
print(f'당신의 총점은{total}이고,\n당신의 평균 점수는{mean}입니다')
```

```python
kor,eng,math = map(int,input('국어, 영어, 수학점수를 /를 기준으로 나눠서 입력하세요 >>').split('/'))
# map(함수, 반복가능 객체)
# 반복가능 객체 내에 있는 각 요소들 마다
# 전달된 함수를 전부 적용시켜주는 함수.
total = kor + eng + math
mean = total/3
print(f'당신의 총점은{total}이고,\n당신의 평균 점수는{mean}입니다')
```

```python
# 산술 연산자 정리
# +,-,*,/,//,%,**
print(10/3) # 나눗셈, 정수끼리 연산해도 실수가 나올 수 있다.
print(10//3) # 몫 연산자
print(10%3) # 나머지 연산자
print(2**3) # 지수 연산자
```
