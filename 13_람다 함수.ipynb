# **1. 람다 함수**

- 단일문으로 표현되는 익명함수
- 코드 상에서 한 번만 사용되는 기능이 있을 때, 굳이 함수로 만들지 않고 일회성으로 만들어 사용
- 함수 또한 클래스를 통해 생성된 객체
- return 형

```python
def square(x):
  return x ** 2
```

```python
print(square(5))
```

```python
#객체
lambda x : x**2
```

```python
# 진짜 함수를 선언한 것이 맞는지 확인해보자!
#익명함수에 이름을 붙여주는 행위!
# 살면서 절대 만나지 않을 모양
test = lambda x : x**2
print(test(6))
```

```python
#이런 식으로도 잘 쓰이지 않는다.
# 다른 함수와 같이 많이 쓰인다.
print((lambda x : x**2)(6))
```

```python
# sorted() : 정렬하는 함수
# key 매개변수. -> 어떤 방식으로 정렬을 할 것인지를 지정
# key 속성에는 대개 함수가 들어간다.
```

```python
datas = ['   watermelon              ', '            apple    ', '           mandarin', 'plum']
```

```python
print(sorted(datas))
```

```python
def func1(data):
  return len(data.strip())
```

```python
# sorted() 사용 시 key 전달 전과 후를 비교
print(sorted(datas))
print(sorted(datas,key = func1))
```

```python
print(sorted(datas,key = lambda data: len(data.strip())))
```

2. map함수
- 각 요소를 주어진 수식에 따라 변형하여 새로운 값을 반환
- 요소의 개수는 차이가 없다.

```python
num = [1, 2, 3, 4, 5, 6, 7, 10, 11, 12, 15, 16]

result = list(map(lambda n : n**2, num))
print(result)
```

3. filter() 함수
- 람다가 유용하게 사용되는 대표적인 함수
- 특정 조건을 만족하는 요소만 남기고 필터링 시킴
- 필터링을 해준다라는 것은 요소의 개수가 달라질 수 있다는 것

```python
num = [1, 2, 3, 4, 5, 6, 7, 10, 11, 12, 15, 16]

result = list(filter(lambda n : n%2==0, num))
print(result)
```

# **4. 콜백 함수**
- 우리가 직접 호출하지 않고 다른 함수에 의해서 호출되는 함수

```python
def wish():
  print('집 가기 10분 전')
```

```python
def callback(func):
  func()
```

```python
callback(wish)
```
