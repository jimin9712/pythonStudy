# **1. 리스트(list)**

- 여러가지 데이터타입의 데이터를 ,(콤마)로 구분하여 [ ]로 감싼 자료구조

```python
# 리스트를 민드는 방법1
list1 = [1,3,5,7,9]
print(list1)
print(type(list1)) # list도 타입이다!

# 리스트를 만드는 방법2
# list() 함수 사용
list2 = list([1,3,5,7,9]) # 객체에 list만 씌우면 list로 변함
print(list2)

#문자만 가지고 만들어보기
list3 = ['이지민', '이유진', '용호중', '이세찬', '박재범']
print(list3)

#여러가지 데이터타입을 섞어서 만들어보기
list4 = [1,3.14,'이지민', True]
print(list4)

# list도 하나의 타입이기 떄문에
# 여러가지 타입을 담을 수 있는 list의 특성 상
# 요소로서 저장할 수 있다.
list5 = [1,3.14,['이지민','배고파']]
print(list5)
list5 = [1,3.14,[list4]]
print(list5)
```

# **2. 리스트의 인덱싱**

```python
list1 = [1,3,5,7,9]
print(list1, '\n')

# 인덱싱하는 방식은 문자열과 동일함
print(list1[2])

# 리스트 인덱싱한 결과는 하나의 독립적인 값으로 봐야한다.
# 그렇기 때문에 데이터타입에 맞게 연산도 가능하다.
print(list1[-1])
```

```python
list5 = [1,3.14,['이지민','배고파']]
print(list5,'\n')

print(list5[2])
print(type(list5[2]))
print(list5[2][1])
```

```python
list6 = [1,3,5,['류호근','홍길동','강감찬',['날','떠','나','지','마']]]

# '나' 뽑기
print(list6[3][3][2])
print(list6[-1][-1][-3])
```

# **3. 리스트의 슬라이싱**

```python
# 슬라이싱도 문자열의 슬라이싱과 똑같다.
list7 = [10,20,30,40,50]
print(list7)

print(list7[:3]) # 0~2까지 불러오기 시작점이 처음부터 일 때는 생략 가능!
print(list7[-3:-1]) # -로 시작하면 무조건 -로 끝나야함 근데 굳이
```

# **4. 리스트 연산자**

```python
list8 = [10,20,30]
list9 = [100,200]

# 리스트끼리의 덧셈 연산은 연결!
print(list8 + list9)
# print(list8 - list9) #리스트의 뺄셈,나눗셈, 곱셈은 안됨. TypeError: unsupported operand type(s) for -: 'list' and 'list'
# print(list8 / list9) TypeError: unsupported operand type(s) for -: 'list' and 'list'
# print(list8 * list9) TypeError: can't multiply sequence by non-int of type 'list'

# 리스트의 정수와 연산은 가능
# 정수만큼 반복해준다.
print(list8 * 3)
```

# **5. 리스트의 수정/삭제**

```python
list9 = [10,20,30]
print(list9)
# 수정
list9[1] = 2222
print(list9)
# 리스트의 수정 슬라이싱은 좀 까다롭다.
```

```python
list10 = [10,20,30]
list11 = [10,20,30]
# 헷갈리면 print 찍어보기
#인덱싱을 사용해서 값을 수정
list11[1] = [55,66,77]
print(list11)
#슬라이싱을 사용해서 값을 수정
list10[1:2]  = [55,66,77] #슬라이싱으로 가져오면 list 타입
print(list10) #[20] 리스트 자리에 55,66,77을 삽입

#위 두 가지의 차이를 한번 눈에 담아주자!
```

```python
# 삭제
list10 = [10,20,30]
list11 = [10,20,30]

# 슬라이싱을 이용한 삭제
list11[1:3] = []
print(list11)

#위의 방식을 인덱싱을 이용해서 한다면?
#값의 수정
#비어있는 리스트도 하나의 값이다.
list10[1] = [] #비어있는 값으로 수정이 되는 것이다.
print(list10)

# del 키워드를 활용한 삭제
del list10[1]
print(list10)

#del과 슬라이싱의 콜라보
list12 = [1,2,3,4,5,6]
del list12[2:5]
print(list12)
```

# **6. 참고 : 얕은 복사와 깊은 복사**
- 얕은 복사 : 주소 복사
- 깊은 복사 : 데이터 복사

```python
#전반적인 프로그래밍 언어의 이해를 위해 C언어 개념이 들어감
#앝은 복사 예시
#이유 : temp1 와 temp2는 같은 주소를 참조하고 있기 때문에
temp1 = [10,20,30,40,50]
temp2 = temp1
# 실제 주소가 아닌 Python에서 관리하는 고유 코드
print('temp1의 주소 : ',id(temp1))
print('temp2의 주소 : ',id(temp2))
temp2[2] = 9999

print('temp1 : ',temp1)
print('temp2 : ',temp2)

# temp1 에 데이터가 담겨있는 것 처럼 보이지만 실제로는 주소값만 가지고 있다.
# 그렇기 때문에 temp2 = temp1 하게 되면 값이 아닌, 해당 주소값을 넣어주는 것이다.
# 그래서 두 변수 모두 같은 주소를 참조하고 있기 때문에 하나의 값만 수정해도 둘 다 바뀐다.
```

```python
# 깊은 복사 예시
temp3 = [1,2,3,4,5,6]
temp4 = temp3.copy()
# a.copy() : 안에 있는 요소를 그대로 복사해서 할당해주는 역할
# 실제 주소가 아닌 Python에서 관리하는 고유 코드
print('temp3의 주소 : ',id(temp3))
print('temp4의 주소 : ',id(temp4))
temp4[2] = 9999

print('temp3 : ',temp3)
print('temp4 : ',temp4)
```

```python
#참고
test = [10,20,30,[40,50,60]]
temp = test.copy()

print('test의 주소', id(test))
print('temp의 주소', id(temp))

print('test의 내부 리스트의 주소', id(test[-1]))
print('temp의 내부 리스트의 주소', id(temp[-1]))

import copy
test2 = [10,20,30,[40,50,60]]
temp2 = copy.deepcopy(test2)

print('test의 주소', id(test2))
print('temp의 주소', id(temp2))

print('test의 내부 리스트의 주소', id(test2[-1]))
print('temp의 내부 리스트의 주소', id(temp2[-1]))

# copy() 함수는 내부 리스트의 요소를 복사해준다
# 리스트의 요소에 리스트가 있다는 것은,
# 값이 아닌 요소가 되는 리스트의 주소가 저장되어 있는 것이다
# 그렇기 때문에 deepcopy()함수를 사용해서,
# 내부 리스트가 얕은 복사되는 것을 방지할 수 있다.
```

# **7. 리스트 함수**

```python
list13 = [10,20,30]
print(list13)

#값 추가하기
#list13[3] = 55 # IndexError: list assignment index out of range
#append() : 리스트 마지막 자리에 데이터를 추가
list13.append(40)
print(list13)
#리스트도 하나의 값이기 때문에, 리스트 타입의 값이 요소로 추가된다.
list13.append([50,60])
print(list13)

#list13.append(10,20) #하나씩만 추가 가능하다!
a = list13 + [10,20]
print(a)
```

```python
#리스트의 삭제
list14 = [10,20,30,40,50]
print(list14)

# pop() : 리스트 마지막에 저장된 데이터를 반환하고 해당 데이터는 삭제
print('삭제된 값 :',list14.pop())
print(list14)

#매개변수로 인덱스를 전달하면 해당 인덱스의 데이터가 삭제
print('삭제된 값 : ',list14.pop(1))
print(list14)
```

```python
list15 = [10,20,30,40,50,20,40,20,20]

#remove() : 매개변수로 전달괸 데이터를 제거
# 같은 데이터가 여러개 있는 경우 처음 출현하는 데이터를 삭제
list15.remove(20)
print(list15)

list15.remove(20)
print(list15)

list15.remove(20)
print(list15)
#삭제할 데이터가 없다면 에러 발생
# list15.remove(20)
# ValueError: list.remove(x): x not in list
```

```python
#리스트의 요소 삽입
list16 = [11,22,33,44]

# lnsert() : 리스트의 특정 위치(인덱스)에 데이터를 삽입

list16.insert(1,49)
print(list16)

list16.insert(-3,77)
print(list16)

# list16.insert(9999) 인덱스 번호는 무조건 전달해줘야한당
#가장 마지막에 추가하고 싶다면
#인덱스 번호 +1을 전달해줘도 된다.
list16.insert(6,55)
print(list16)
```

```python
#리스트의 정렬
list17 = [10,40,30,100,90,50,20]
#sort() : 리스트에 저장된 데이터를 오름차순, 내림차순으로 정렬
#내가 접근한 리스트를 정렬하고 수정까지 !
list17.sort() #오름차순
print(list17)

#앞 뒤 역전
# list17.reverse()
# print(list17)

list17.sort(reverse = True) #내림차순
print(list17)

list18 = ['Apple', 'banana', 'Banana', 'watermelon', 'orange']
list18.sort()
print(list18)

list19 = ['용호중', '정우성', '강동원', '원빈', '차은우']
list19.sort()
print(list19)
```

```python
list20 = ['용호중', '정우성', '강동원', '원빈', '차은우']

#sorted() 원본데이터는 수정하지 않고 정렬된 리스트를 반환
print(sorted(list20))
print(list20) # 원본 데이터는 그대로!

#정렬된 리스트를 계속 사용하고 싶다면(변수)에 저장한다
res = sorted(list20)
print(res)
```
